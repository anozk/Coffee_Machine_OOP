```python
# Solved stage 6/6 Object-oriented programming

money = 550
water = 400
milk = 540
coffee_beans = 120
cups = 9

cappuccino_water = 200
cappuccino_milk = 100
cappuccino_coffee_beans = 12
cappuccino_money = 6
cappuccino_cups = 1 

espresso_water = 250
espresso_coffee_beans = 16
espresso_money = 4
espresso_cups = 1 

latte_water = 350
latte_milk = 75
latte_coffee_beans = 20
latte_money = 7
latte_cups = 1 

fill = "fill"
take = 'take'
buy = 'buy'
remaining = 'remaining'
exit = 'exit'
action_water = 'action_water', 
action_milk =  "action_milk", 
action_coffe_beans = "action_coffe_beans"
action_cups = "action_cups"
action = "action"


class CoffeeMachine:
  def __init__(self, money, water, milk, coffee_beans, cups, espresso_water, espresso_coffee_beans, espresso_cups, latte_water,
               latte_milk, latte_coffee_beans, latte_money, latte_cups, cappuccino_water, cappuccino_milk, cappuccino_coffee_beans,
               cappuccino_money, cappuccino_cups):
    self.money = money
    self.water = water
    self.milk = milk
    self.coffee_beans = coffee_beans
    self.cups = cups
    self.espresso_water = espresso_water
    self.espresso_water = espresso_coffee_beans
    self.espresso_money = espresso_money
    self.espresso_cups = espresso_cups
    self.latte_water = latte_water
    self.latte_milk = latte_milk
    self.latte_coffee_beans = latte_coffee_beans
    self.latte_money = latte_money
    self.latte_cups = latte_cups
    self.cappuccino_water = cappuccino_water
    self.cappuccino_milk = cappuccino_milk
    self.cappuccino_coffee_beans = cappuccino_coffee_beans
    self.cappuccino_money = cappuccino_money
    self.cappuccino_cups = cappuccino_cups


  def choose_cooffee(self, action, action_water, action_milk, action_coffe_beans, action_cups):
    while True:
      action = input("Write action (buy, fill, take, remaining, exit): ")
      if action == "fill":
        action_water = int(input("Write how many ml of water you want to add: "))
        action_milk = int(input("Write how many ml of milk you want to add: "))
        action_coffe_beans = int(input("Write how many grams of coffee beans you want to add: "))
        action_cups = int(input("Write how many disposable coffee cups you want to add: "))
        self.water += action_water
        self.milk += action_milk
        self.coffee_beans += action_coffe_beans
        self.cups += action_cups
      elif action == 'buy':
        action1 = input("What do you want to buy? 1 - espresso, 2 - latte, 3 - cappuccino, back - to main menu: ")
        if action1 == "1":
          if self.water >= espresso_water and self.coffee_beans >= espresso_coffee_beans and self.cups >= espresso_cups:
            print("I have enough resources, making you a coffee!")
            self.water = self.water - espresso_water
            self.coffee_beans = self.coffee_beans - espresso_coffee_beans
            self.milk = self.milk # made a change here
            self.cups = self.cups - espresso_cups 
            self.money = self.money + espresso_money
          elif self.water < espresso_water: 
            print("Sorry, not enough water!")
          elif self.coffee_beans < espresso_coffee_beans:
            print("Sorry, not enough coffee beans!")
          elif self.cups < espresso_cups:
            print("Sorry, not enough cups!")
        elif action1 == "2":
          if self.water >= latte_water and self.milk >= latte_milk and self.coffee_beans >= latte_coffee_beans and self.cups >= latte_cups:
            print("I have enough resources, making you a coffee!")
            self.water = self.water - latte_water
            self.milk = self.milk - latte_milk
            self.coffee_beans = self.coffee_beans - latte_coffee_beans
            self.cups = self.cups - latte_cups
            self.money = self.money + latte_money
          elif self.water < latte_water:
            print("Sorry, not enough water!")
          elif self.milk < latte_milk:
            print("Sorry, not enough milk!")
          elif self.coffee_beans < latte_coffee_beans:
            print("Sorry, not enough coffee beans!")
          elif self.cups < latte_cups:
            print("Sorry, not enough cups!")
        elif action1 == "3":
          if self.water >= cappuccino_water and self.milk >= cappuccino_milk and self.coffee_beans >= cappuccino_coffee_beans and self.cups >= cappuccino_cups:
            print("I have enough resources, making you a coffee!")
            self.water = self.water - cappuccino_water 
            self.milk = self.milk - cappuccino_milk 
            self.coffee_beans = self.coffee_beans - cappuccino_coffee_beans
            self.cups = self.cups - cappuccino_cups 
            self.money = self.money + cappuccino_money
          elif self.water < cappuccino_water:
            print("Sorry, not enough water!")
          elif self.milk < cappuccino_milk:
            print("Sorry, not enough milk!")
          elif self.coffee_beans < cappuccino_coffee_beans:
            print("Sorry, not enough coffee beans!")
          elif self.cups < cappuccino_cups:
            print("Sorry, not enough cups!")
        elif action1 == 'back':
            pass
      elif action == 'remaining':
        self.water = self.water
        self.milk = self.milk
        self.coffee_beans = self.coffee_beans 
        self.cups = self.cups
        self.money = self.money
        print("The coffee machine has:\n", self.water, "ml of water\n", self.milk, "ml of milk\n", self.coffee_beans, "g of coffee beans\n",
              self.cups, "disposable cups\n", "$"+str(self.money), "of money")
      elif action == 'take':
         print("I gave you: " + "$"+str(self.money))
         self.money = self.money - self.money
      elif action == 'exit':
        print('exit')
        break
      else:
        print('choose another action')

action_ = CoffeeMachine(money, water, milk, coffee_beans, cups, espresso_water, espresso_coffee_beans, espresso_cups, latte_water,
               latte_milk, latte_coffee_beans, latte_money, latte_cups, cappuccino_water, cappuccino_milk, cappuccino_coffee_beans,
               cappuccino_money, cappuccino_cups)

action_.choose_cooffee(action, action_water, action_milk, action_coffe_beans, action_cups)
```
